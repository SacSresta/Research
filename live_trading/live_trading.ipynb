```python
from alpaca.trading.client import TradingClient
from alpaca.trading.requests import GetAssetsRequest
from alpaca.trading.enums import AssetClass

search_params = GetAssetsRequest(asset_class=AssetClass.US_EQUITY)

assets = trading_client.get_all_assets(search_params)
```

```python
# Convert each Asset object to a dictionary using `.dict()` or `.model_dump()` (depending on the library version)
asset_dicts = [asset.model_dump() for asset in assets]  # or asset.model_dump() for newer versions of Pydantic

# Now convert to DataFrame
df = pd.DataFrame(asset_dicts)
```

```python
from alpaca.trading.stream import TradingStream

trading_stream = TradingStream(api, secret, paper=True)

async def update_handler(data):
    # trade updates will arrive in our async handler
    print(data)

# subscribe to trade updates and supply the handler as a parameter
trading_stream.subscribe_trade_updates(update_handler)

# start our websocket streaming
trading_stream.run()
```

```python
from alpaca.data.historical import NewsClient
from alpaca.data.requests import NewsRequest
from datetime import datetime

# Initialize the NewsClient
news_client = NewsClient(api, secret,raw_data=True)

# Create a NewsRequest object with the required parameters
news_request = NewsRequest(
    symbol_or_symbols= ['TSLA','AAPL','AMD'], limit = 40
)
# Fetch the news data
news_data = news_client.get_news(request_params=news_request)

```
