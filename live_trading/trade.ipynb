```python
import alpaca_trade_api as tradeapi

API_KEY = 'PKY7ODAXYMWTX4JK7DVR'
API_SECRET = 'yvQclRacekxCgwL0KX70SFkdbBAsVRvrBQnYaYGY'
BASE_URL = 'https://paper-api.alpaca.markets'
api = tradeapi.REST(API_KEY, API_SECRET, BASE_URL, api_version='v2')

# Parameters
symbol = 'AAPL'
buy_price_threshold = 170
qty_to_buy = 1

# Get current market price
barset = api.get_bars(symbol, tradeapi.TimeFrame.Minute, limit=1000)
current_price = barset[0].c  # close price of last minute

print(f"Current {symbol} price: ${current_price}")

# Trading logic
if current_price < buy_price_threshold:
    print(f"Buying {qty_to_buy} share(s) of {symbol} at ${current_price}")
    api.submit_order(
        symbol=symbol,
        qty=qty_to_buy,
        side='buy',
        type='market',
        time_in_force='gtc'
    )
else:
    print(f"No trade. {symbol} price is above ${buy_price_threshold}")
```

```python
import datetime as dt
# Define time range - explicitly requesting 60 days
end_date = dt.datetime.now()
start_date = end_date - dt.timedelta(days=60)

# Format dates as strings in YYYY-MM-DD format
start_str = start_date.strftime('%Y-%m-%d')
end_str = end_date.strftime('%Y-%m-%d')

# Request with explicit timeframe
barset = api.get_bars(
    symbol, 
    tradeapi.TimeFrame.Day, 
    start=start_str,
    end=end_str,
    feed='iex'
)
```

```python
data = [bar.__dict__ if hasattr(bar, '__dict__') else bar for bar in barset]
df = pd.DataFrame(data)
data_dicts= df['_raw'].tolist()
df = pd.DataFrame(data_dicts)
df['t'] = pd.to_datetime(df['t'])
df = df.rename(columns={
    'c': 'Close',
    'h': 'High',
    'l': 'Low',
    'n': 'Trades',
    'o': 'Open',
    't': 'Date',  
    'v': 'Volume',
    'vw': 'VWAP' 
}).set_index('Date')

df
```
