```python
import re
def parse_model_params(params_str):
    # Regular expression to match model names and their parameters
    pattern = r"'([^']+)': ([\w.() ,]+)"
    matches = re.findall(pattern, params_str)

    # Create a dictionary to hold the model instances
    model_dict = {}
    
    for model_name, params in matches:
        model_dict[model_name] =model_name + '(' + params + ')'

    return model_dict
```
