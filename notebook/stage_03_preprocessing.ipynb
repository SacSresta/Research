```python
# Install required libraries
# pip install transformers nltk textblob vaderSentiment

# Sentiment analysis using FinBERT (financial-domain specific BERT model)
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np

# Load FinBERT model
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")

# Function to get sentiment scores
def get_finbert_sentiment(headlines):
    sentiments = []
    for headline in headlines:
        inputs = tokenizer(headline, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)
        
        # Get probabilities
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1).detach().numpy()[0]
        # FinBERT classes: negative (0), neutral (1), positive (2)
        sentiment_score = probs[2] - probs[0]  # positive - negative
        sentiments.append(sentiment_score)
    
    return np.array(sentiments)

```

```python
# 1. MACD Signal
df['macd_signal_line'] = np.where(
    (df['macd'] > df['macd_signal']) & (df['macd_hist'] > 0), 1,    # Bullish
    np.where((df['macd'] < df['macd_signal']) & (df['macd_hist'] < 0), -1, 0)  # Bearish
)

# 2. RSI Signal
df['rsi_signal'] = np.where(
    df['rsi'] < 30, 1,    # Oversold - Buy
    np.where(df['rsi'] > 70, -1, 0)  # Overbought - Sell
)

# 3. VWAP Signal
df['vwap_signal'] = np.where(
    df['Close'] > df['vwap'], 1,    # Price above VWAP - Bullish
    -1                              # Price below VWAP - Bearish
)

# 4. Bollinger Bands Signal
df['bb_signal'] = np.where(
    df['Close'] < df['bbands_lower'], 1,    # Price below lower band - Potential buy
    np.where(df['Close'] > df['bbands_upper'], -1, 0)  # Price above upper band - Potential sell
)

# 5. Combined Signal (Simple Average)
df['combined_signal'] = (
    df['macd_signal_line'] + 
    df['rsi_signal'] + 
    df['vwap_signal'] + 
    df['bb_signal']
) / 4

# 6. Trend Strength
df['trend_strength'] = np.where(
    (df['Close'] > df['sma_20']) & (df['sma_20'] > df['sma_50']), 1,    # Strong uptrend
    np.where((df['Close'] < df['sma_20']) & (df['sma_20'] < df['sma_50']), -1, 0)  # Strong downtrend
)

# 7. Volume-Based Signal
df['volume_signal'] = np.where(
    (df['volume_momentum'] > 0) & (df['price_momentum'] > 0), 1,    # Rising volume with rising price
    np.where((df['volume_momentum'] > 0) & (df['price_momentum'] < 0), -1, 0)  # Rising volume with falling price
)

# 8. Final Trading Signal (Weighted Combination)
df['trading_signal'] = (
    0.3 * df['macd_signal_line'] +
    0.2 * df['rsi_signal'] +
    0.2 * df['vwap_signal'] +
    0.15 * df['trend_strength'] +
    0.15 * df['volume_signal']
)

# 9. Signal Strength Categories
df['signal_strength'] = pd.cut(
    df['trading_signal'],
    bins=[-1, -0.5, -0.2, 0.2, 0.5, 1],
    labels=['Strong Sell', 'Moderate Sell', 'Neutral', 'Moderate Buy', 'Strong Buy']
)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.svm import SVC

# Create features
def create_features(df):
    features = pd.DataFrame()
    
    # Sentiment features
    features['sentiment_score'] = df['sentiment_score']
    features['sentiment_ma5'] = df['sentiment_score'].rolling(5).mean()
    features['sentiment_ma10'] = df['sentiment_score'].rolling(10).mean()
    features['sentiment_std5'] = df['sentiment_score'].rolling(5).std()
    
    # VWAP signal features
    features['shifted_vwap_signal'] = df['shifted_vwap_signal']
    features['vwap_signal_ma5'] = df['shifted_vwap_signal'].rolling(5).mean()

    
    return features.dropna()
```

```python

# Prepare data
features_df = create_features(data)
X = features_df.drop('shifted_vwap_signal', axis=1)
y = features_df['shifted_vwap_signal']
```

```python
# Time series split
tscv = TimeSeriesSplit(n_splits=5)
```

```python

# Models to test
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(probability=True, random_state=42)
}
```

```python
# Train and evaluate models
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    fold_scores = []
    
    for fold, (train_idx, test_idx) in enumerate(tscv.split(X)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        # Scale features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Train and predict
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        # Store results
        fold_scores.append({
            'classification_report': classification_report(y_test, y_pred),
            'confusion_matrix': confusion_matrix(y_test, y_pred)
        })
        
        print(f"\nFold {fold + 1} Results:")
        print("Classification Report:")
        print(fold_scores[-1]['classification_report'])
        print("\nConfusion Matrix:")
        print(fold_scores[-1]['confusion_matrix'])
    
    results[name] = fold_scores
```

```python
# Feature importance for Random Forest
rf_model = models['Random Forest']
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_model.feature_importances_
})
print("\nRandom Forest Feature Importance:")
print(feature_importance.sort_values('importance', ascending=False))
```

```python
# Best model selection
best_model = None
best_score = -1
for name, fold_scores in results.items():
    avg_score = np.mean([float(score['classification_report'].split('\n')[-2].split()[-2]) for score in fold_scores])
    if avg_score > best_score:
        best_score = avg_score
        best_model = name

print(f"\nBest performing model: {best_model} with average score: {best_score:.3f}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import LabelEncoder
import numpy as np
from transformers import BertTokenizer
```

```python

# Custom Dataset
class HeadlineDataset(Dataset):
    def __init__(self, headlines, labels, tokenizer, max_length=128):
        self.headlines = headlines
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
        
    def __len__(self):
        return len(self.headlines)
    
    def __getitem__(self, idx):
        headline = str(self.headlines[idx])
        encoding = self.tokenizer(
            headline,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(self.labels[idx], dtype=torch.long)}
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(data['shifted_vwap_signal'])
encoded_labels
```

```python
# Model parameters
VOCAB_SIZE = tokenizer.vocab_size
EMBEDDING_DIM = 100
HIDDEN_DIM = 256
OUTPUT_DIM = len(label_encoder.classes_)
N_LAYERS = 2
DROPOUT = 0.5
BATCH_SIZE = 32
N_EPOCHS = 10
```

```python
# Time series split
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# LSTM Model
class LSTMSentiment(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, n_layers, dropout):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, n_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, input_ids, attention_mask):
        embedded = self.embedding(input_ids)
        packed_output, (hidden, cell) = self.lstm(embedded)
        hidden = self.dropout(hidden[-1])
        output = self.fc(hidden)
        return output
```

```python
# Training function
def train_model(model, train_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    
    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)
        
        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    return total_loss / len(train_loader)
```

```python
# Evaluation function
def evaluate_model(model, test_loader, criterion, device):
    model.eval()
    total_loss = 0
    predictions = []
    actual = []
    
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)
            
            outputs = model(input_ids, attention_mask)
            loss = criterion(outputs, labels)
            
            total_loss += loss.item()
            predictions.extend(outputs.argmax(dim=1).cpu().numpy())
            actual.extend(labels.cpu().numpy())
    
    return total_loss / len(test_loader), predictions, actual
```

```python
for fold, (train_idx, test_idx) in enumerate(tscv.split(data)):
        print(f"\nFold {fold + 1}")
        
        # Create datasets
        train_dataset = HeadlineDataset(
            df['headline'].iloc[train_idx],
            encoded_labels[train_idx],
            tokenizer
        )
        test_dataset = HeadlineDataset(
            df['headline'].iloc[test_idx],
            encoded_labels[test_idx],
            tokenizer
        )
        
        # Create dataloaders
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
        
        # Initialize model
        model = LSTMSentiment(
            VOCAB_SIZE, EMBEDDING_DIM, HIDDEN_DIM, 
            OUTPUT_DIM, N_LAYERS, DROPOUT
        ).to(device)
        
        # Loss and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = torch.optim.Adam(model.parameters())
        
        # Training loop
        for epoch in range(N_EPOCHS):
            train_loss = train_model(model, train_loader, optimizer, criterion, device)
            test_loss, predictions, actual = evaluate_model(model, test_loader, criterion, device)
            
            print(f'Epoch: {epoch+1}')
            print(f'Train Loss: {train_loss:.4f}')
            print(f'Test Loss: {test_loss:.4f}')
            
        # Final evaluation
        _, predictions, actual = evaluate_model(model, test_loader, criterion, device)
        print("\nClassification Report:")
        print(classification_report(actual, predictions))
```

```python
# Function to get sentiment scores
def get_finbert_sentiment(headlines):
    sentiments = []
    for headline in headlines:
        inputs = tokenizer(headline, return_tensors="pt", padding=True, truncation=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)
        
        # Get probabilities
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1).detach().numpy()[0]
        # FinBERT classes: negative (0), neutral (1), positive (2)
        sentiment_score = probs[2] - probs[0]  # positive - negative
        sentiments.append(sentiment_score)
    
    return np.array(sentiments)
```
